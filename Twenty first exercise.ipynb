```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('3.12. Example.csv')
data
```

```python
plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel('Satisfcation')
plt.ylabel('Loyality')
plt.show()
```

```python
x = data.copy()
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['clusters_pred'] = kmeans.fit_predict(x)
clusters
```

```python
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c = clusters['clusters_pred'], cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
from sklearn import preprocessing 
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
number_of_clusters = range(1,11)
```

```python
plt.plot(number_of_clusters, wcss)
plt.xlabel('Number of clusters')
plt.ylabel('Within-Clusters Sum of Squares')
plt.show()
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
```

```python
clusters_new = x.copy()
clusters_new['Cluster'] = kmeans_new.fit_predict(x_scaled)
clusters_new
```

```python
plt.scatter(clusters_new['Satisfaction'], clusters_new['Loyalty'], c = clusters_new['Cluster'], cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```
